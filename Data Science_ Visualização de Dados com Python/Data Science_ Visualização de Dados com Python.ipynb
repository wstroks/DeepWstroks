### 1 - Visualização de dados

```
import matplotlib.pyplot as plt
import numpy as np

```

```
x = [1,2]
y = [2,3]

plt.title("Dados Python")

plt.xlabel("Eixo x")

plt.ylabel("Eixo y")
plt.plot(x,y)
plt.show()

```

```
x = [1,3,5,7,9]
y = [2,3,7,1,0]

x1 = [2,4,6,8,10]
y1= [5,1,3,7,4]

plt.title("Dados Python")

plt.xlabel("Eixo x")

plt.ylabel("Eixo y")

plt.bar(x,y, label = "Grupo 1")
plt.bar(x1,y1,label="Grupo 2")
plt.legend()
plt.show()
```

```
x = [1,3,5,7,9]
y = [2,3,7,1,0]



plt.title("Dados Python")

plt.xlabel("Eixo x")

plt.ylabel("Eixo y")

plt.scatter(x,y, label = "Meus pontos",color="red")
plt.plot(x,y,color="red")


plt.legend()
plt.show()
```

```
x = [1,3,5,7,9]
y = [2,3,7,1,0]



plt.title("Dados Python")

plt.xlabel("Eixo x")

plt.ylabel("Eixo y")


plt.plot(x,y,color="red")
plt.scatter(x,y, label = "Meus pontos",color="red", marker="h",s=100)


plt.legend()
plt.show()
```

```
x = [1,3,5,7,9]
y = [2,3,7,1,0]



plt.title("Dados Python")

plt.xlabel("Eixo x")

plt.ylabel("Eixo y")


plt.plot(x,y,color="k",linestyle=":")
plt.scatter(x,y, label = "Meus pontos",color="r", marker="x",s=100)


plt.legend()
plt.show()
```

```
x = [1,3,5,7,9]
y = [2,3,7,1,0]



plt.title("Dados Python")

plt.xlabel("Eixo x")

plt.ylabel("Eixo y")


plt.plot(x,y,color="k",linestyle=":")
plt.scatter(x,y, label = "Meus pontos",color="r", marker=".",s=100)


plt.legend()
plt.show()
```

```
x = [1,3,5,7,9]
y = [2,3,7,1,0]



plt.title("Dados Python")

plt.xlabel("Eixo x")

plt.ylabel("Eixo y")


plt.plot(x,y,color="m",linestyle="--")
plt.scatter(x,y, label = "Meus pontos",color="r", marker=".",s=100)


plt.legend()
plt.show()
plt.savefig("n.png",dpi=150)
```
